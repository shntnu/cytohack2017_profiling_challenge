```python
import numpy as np
import pandas as pd
import seaborn as sb
import glob
%matplotlib inline
```

```python
files = glob.glob('*augmented.csv')
```

```python
data = {}
for f in files:
    data[f[:5]] = pd.read_csv(f)
```

```python
fullframe = pd.concat(data.values())
```

```python
smiles = fullframe['Metadata_smiles'].unique()
```

```python
featcols = [c for c in fullframe.columns if c[:4] != 'Meta']
```

```python
metacols = [c for c in fullframe.columns if c[:4] == 'Meta']
```

```python
dmso = fullframe[fullframe['Metadata_broad_sample_type'] != 'trt']
dfeat = dmso[featcols]
dmean = dfeat.mean()
dstd = dfeat.std()
temp = []
for smile in smiles:    
    cframe = fullframe[fullframe['Metadata_smiles'] == smile]
    metaframe = cframe[metacols]
    metaframe = metaframe[1:2]
    featframe = cframe[featcols]
    featframe = (featframe - dmean) / dstd
    featframe = featframe.mean(level=0)
    newframe = metaframe.join(featframe)
    temp.append(newframe)
```

```python
meanframe = pd.concat(temp)
meanframe.sort(columns=["Metadata_smiles"], inplace=True)
meanframe
```

```python
dropframe = meanframe.dropna(axis=1)
```

```python
featdropcols = [c for c in dropframe.columns if c[:4] != 'Meta']
```

```python
cormatr = np.corrcoef(np.asarray(dropframe[featdropcols]))
```

```python
np.savetxt('cormatr.csv',cormatr,delimiter=',')
```

```python
fullframe.to_csv('fullframe.csv')
```

```python
meanframe.to_csv('meanframe.csv')
```
